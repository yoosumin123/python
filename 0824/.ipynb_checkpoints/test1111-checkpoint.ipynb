```python
#1. 테이블 완성
import pandas as pd
import numpy as np
df = pd.DataFrame({'id':['1','2','3','4','5'],
                  'kuk':[90,70,40,80,50],
                  'eng':[100,80,60,70,30],
                  'mat':[80,90,50,90,70]
                  })
df
```

```python
# 2.수학 점수가 90점 이상인 학생들만 출력

print(df['mat']>=90)
```

```python
#3. 총점 필드 추가
df['총점']=(df['kuk']+df['eng']+df['mat'])
df
```

```python
# 4-1.평균 필드 추가 
df['평균']=(df['kuk']+df['eng']+df['mat'])/3
df
```

```python
# 4-2. 내림차순 정렬
df.sort_values('평균',ascending=False)
```

```python
# 5. 평균 점수가 60미만이면 결측치(NA)를 입력
df['평균']=(df['kuk']+df['eng']+df['mat'])/3
df.iloc[3,5] = np.nan
df.iloc[4,5] = np.nan
df
```
